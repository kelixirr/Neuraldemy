#### Difference between Series and Numpy array 

A key difference between Series and ndarray is that operations between Series automatically align the data based on label. Thus, you can write computations without giving consideration to whether the Series involved have the same labels.

```python
s = pd.Series(np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'])
s
```

```python
s[1:]    #creating one series using indexing - don't worry we will study indexing of series later on 
```

```python
s[:-1]    # creating another series using indexing 
```

```python
s[1:] + s[:-1]   # adding to separate unaligned series 
```

Notice the operation here is happening elementwise from same label in one series to another. We are using the same series to slice into two differnt series so don't get confuse. 

The result of an operation between unaligned Series will have the union of the indexes involved. If a label is not found in one Series or the other, the result will be marked as missing NaN. Being able to write code without doing any explicit data alignment grants immense freedom and flexibility in interactive data analysis and research. The integrated data alignment features of the pandas data structures set pandas apart from the majority of related tools for working with labeled data.

```python
s = pd.Series(np.random.randn(5), name="something") # Series also has a name attribute 
s
```

2-D numpy.ndarray

Structured or record ndarray

A Series

Another DataFrame

If axis labels are not passed, they will be constructed from the input data based on common sense rules.

## Working with missing data 
This time our data does not have any missing value but in real world projects we will have missing values so we need to keep in mind how to deal with them'. 
You can use fillna() function to fill the missing values using the mean or media of that column and use inplace=True to update the value in the actual data. By deafault the keyword argument inplace is set to False so it returns a copy of the datafram after dealing with missing data. So make sure to reassign it to a new variable if you want to work with non-missing data without affecting the actual data. In case, you want to update the actual data at the same time use inplace = True. We will see examples of missing data in future lectures. 

__
 
In case you want to drop the rows with missing data you can use dropna(). This removes rows with missing value. Addiotnally this also accepts inplace keyword argument which is set to false by default so make sure to set it to True if you want to update you data directly.
